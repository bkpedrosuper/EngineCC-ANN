# Gather Data

## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', 
                  font_scale=1, color_codes=True, rc=None)
```

## Get Data

```python
data_path = '../data/MotorUniversal_2k_5kHz_SemNorm.xlsx'
data = pd.read_excel(data_path)

data.head()
```

### Transform columns

```python
columns = ['state'] + [f'current_{i}' for i in range(1, 1001)] + [f'volt_{i}' for i in range(1, 1001)]
columns

data.columns = columns

data
```

### Info about raw Dataset

#### Shape

```python
data.shape
```

#### Important info

```python
data_description = data.describe()

df_filters = data.filter(items=['state', 'current_500', 'volt_500'])

df_filters.describe()
```

```python
for column in df_filters.columns:
    sns.histplot(data=df_filters, hue='state', x=column)
    plt.show()
    plt.clf()
```
