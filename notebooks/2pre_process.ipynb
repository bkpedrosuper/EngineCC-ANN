## Handle Outliers
Lets handle the outliers using the Inter Quantile Range method:

```python
def removeOutliers(data, col):
    Q3 = np.quantile(data[col], 0.75)
    Q1 = np.quantile(data[col], 0.25)
    IQR = Q3 - Q1

    global outlier_free_list
    global filtered_data

    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    outlier_free_list = [x for x in data[col] if (
        (x > lower_range) & (x < upper_range))]
    filtered_data = data.loc[data[col].isin(outlier_free_list)]
```

Handle outliers in current

```python
for column in data.iloc[:, :1000].columns:
    removeOutliers(data, column)
data.iloc[:, :1000]
```

Handle outliers in voltage

```python
for column in data.iloc[:, 1001:2001].columns:
    removeOutliers(data, column)
data.iloc[:, 1000:2000]
```

## Normalization

Lets normalize the voltage data into values from 0 to 1
There is no need to normalize current values as its values are between 0.62 and 0.93	

```python
from sklearn.preprocessing import MinMaxScaler

X_volt = data.iloc[:, 1000:2000]
X_volt
norm = MinMaxScaler()
norm.fit(X_volt)

X_norm = norm.transform(X_volt)

# Now, save the normalized data into the dataset
data.iloc[:, 1000:2000] = X_norm

data
```

The normalization process showed that the characteristic behaviour of each state wasn't lost in the process.
