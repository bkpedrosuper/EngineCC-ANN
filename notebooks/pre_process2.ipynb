## Normalization

Lets normalize the voltage data into values from 0 to 1
There is no need to normalize current values as its values are between 0.62 and 0.93	

```python
from sklearn.preprocessing import MinMaxScaler

X_volt = data.iloc[:, 1000:2000]
X_volt
norm = MinMaxScaler()
norm.fit(X_volt)

X_norm = norm.transform(X_volt)
```

Now, save the normalized data into the dataset

```python
# Update the values into the dataset
data.iloc[:, 1000:2000] = X_norm

data
```

The normalization process showed that the characteristic behaviour of each state wasn't lost in the process.

## Selecting interesting data
Now, lets get only the current and voltage from measurements from 200 to 800

```python
from config import END_COL, INITIAL_COL
indexes = data.index
states = data['state']

# drop current
data_curr = data.iloc[:, 1+INITIAL_COL:1+END_COL]
data_curr

#drop voltage
data_volt = data.iloc[:, 1001+INITIAL_COL:1001+END_COL]
data_volt

data = pd.concat([data_curr, data_volt], ignore_index=False, axis=1)

data['state'] = states

data
```
