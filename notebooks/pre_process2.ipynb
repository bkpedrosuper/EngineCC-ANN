# Pre Processing

With the information gathered from the previous session, we can now pre-process our data.

```python
import pandas as pd
```

Read data

```python
data_path = '../data/dataset01.csv'
data = pd.read_csv(data_path)
data
```

As we only got a few missed values, lets remove them

```python
data = data.dropna()

data.head()
```

Now, lets get only the current and voltage from measurements from 200 to 800

```python
from config import END_COL, INITIAL_COL
indexes = data.index
states = data['state']

# drop current
data_curr = data.iloc[:, 1+INITIAL_COL:1+END_COL]
data_curr

#drop voltage
data_volt = data.iloc[:, 1001+INITIAL_COL:1001+END_COL]
data_volt

data = pd.concat([data_curr, data_volt], ignore_index=False, axis=1)

data['state'] = states

data
```

## Normalization

Lets normalize the data into values from 0 to 1

```python
from sklearn.preprocessing import MinMaxScaler

X, y = data.iloc[:, :-1], data.iloc[:, -1]

norm = MinMaxScaler()
norm.fit(X)

X_norm = norm.transform(X)
```

Now, save the normalized data into the dataset

```python
# Update the values into the dataset
data.iloc[:, :-1] = X_norm

data
```

The normalization process showed that the characteristic behaviour of each state was lost in the process.

Agora, vamos salvar o dataset pre-processado em um arquivo csv

```python
data.to_csv('../data/dataset_cleaned.csv', index=False)
```
